```python
# 10809 알파벳 찾기
# 알파벳은 26개!
from string import ascii_lowercase
alphabet = list(ascii_lowercase)
answer = [-1 for x in range(26)]
problem = [x for x in input()]

order =0    
cnt = 0

for i in problem :
    for j in alphabet :
        if j == i :
            if answer[cnt] != -1 :
                order +=1
            else :
                answer[cnt] = order 
                order += 1

        else : 
            cnt +=1    
    cnt = 0


answer =' '.join(s for s in [str(y) for y in answer])

print(answer)
```

```python
# 2675 문자열 반복
T = int(input())
for i in range(T):
    a =input().split()
    R = int(a[0])
    P=''
    for x in a[1]:
        P = P+x*R
    print(P)
```

```python
# 1157 단어 공부
a = input().upper()
s = [x for x in a]
p = list(set(s))
cnt = []

for i in p :
    b = s.count(i)
    cnt.append(b)

order = sorted(cnt, reverse = True)

if cnt.count(order[0]) == 1 :
    answer = p[cnt.index(max(cnt))]
else :
    answer = '?'
print(answer)
```

```python
# 2908 상수
a,b = map(str, input().split())
A = [x for x in a]
A = list(reversed(A))
A = ''.join(A)
B = [x for x in b]
B = list(reversed(B))
B = ''.join(B)
if int(A) > int(B) :
    print(A)
else :
    print(B)
```

```python
# 1316 그룹단어체커
n = int(input())
cnt =0
for i in range(n) :
    word = input()
    for j in range(len(word)-1) :
        if word[j] != word[j+1] :
            check = word[j+2:]
            if word[j] in check :
                cnt -= 1
                break
    cnt+=1
print(cnt)
```
